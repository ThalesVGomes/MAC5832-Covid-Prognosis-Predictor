# Exploratory Data Analysis - Covid Prognosis Dataset

---

### 1) Setup

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot params
sns.set()
plt.rcParams["figure.figsize"] = (15,8)
```

```python
# Pandas view params
pd.set_option("display.max_columns", None)
```

```python
TRAIN_CLINICAL_FILENAME = "train_set_clinical.csv"
TEST_CLINICAL_FILENAME = "test_set_clinical.csv"
```

---

### 2) Read Data

```python
train = pd.read_csv(TRAIN_CLINICAL_FILENAME, sep=";", index_col="ID")
test = pd.read_csv(TEST_CLINICAL_FILENAME,  sep=";", index_col="ID")
```

```python
train.shape, test.shape
```

```python
train.head()
```

```python
test.head()
```

```python
label_col = ["Group"]
protein_cols = [col for col in train.columns if "Freq" in col]
clinical_cols = list(set(train.columns) - set(protein_cols + label_col))
symptoms_cols = [col for col in clinical_cols if col not in ["Age (Y)", "Sex"]]
```

---

### 3) Exploration

```python
# Check missing values - TRAIN
train.isna().sum().to_frame().sort_values(0, ascending=False).head()
```

```python
train[train.isna().any(1)]
```

```python
# Check missing values - TEST
test.isna().sum().to_frame().sort_values(0, ascending=False).head()
```

- Temos apenas um caso com dados faltantes no set de treino, sendo que estes são apenas os dados clínicos
- No set de teste não há dados faltantes

```python
# Check values == 0 - TRAIN
train_zeros = np.sum(train==0).to_frame().sort_values(0, ascending=False).rename({0:"qt_0"}, axis=1)
train_zeros["%_qt_0"] = train_zeros["qt_0"]/train.shape[0]
train_zeros.head(20)
```

```python
# Check missing values - TEST
test_zeros = np.sum(test==0).to_frame().sort_values(0, ascending=False).rename({0:"qt_0"}, axis=1)
test_zeros["%_qt_0"] = test_zeros["qt_0"]/test.shape[0]
test_zeros.head(20)
```

- Existem features com grande quantidades de zeros (algumas mais da metade)
- A maioria dos casos acima representam as features de sintomas, o que tem sentido se levado em consideração a menor quantidade de casos severos no dataset

```python
# Label Distribution - TRAIN
train_labels = train.Group.value_counts().reset_index().rename({"index":"label", "Group":"qt"}, axis=1)
train_labels["%_label"] = train_labels.qt/train_labels.qt.sum()
train_labels
```

```python
# Label Distribution - TEST
test_labels = test.Group.value_counts().reset_index().rename({"index":"label", "Group":"qt"}, axis=1)
test_labels["%_label"] = test_labels.qt/test_labels.qt.sum()
test_labels
```

- A proporção de labels entre datasets está diferente, mas os valores são próximos

```python
# Clinical Features Distribution
for col in clinical_cols:
    if col != "Sex":
        fig, ax = plt.subplots(ncols=2)
        _ = sns.histplot(train[col], ax=ax[0])
        _ = sns.histplot(test[col], ax=ax[1])
        ax[0].set_title("TRAIN")
        ax[1].set_title("TEST")
```

```python
# Clinical Features Distribution - Sex
fig, ax = plt.subplots(ncols=2)
train["Sex"].value_counts().plot(kind="bar", ax=ax[0])
test["Sex"].value_counts().plot(kind="bar", ax=ax[1])
ax[0].set_title("TRAIN")
ax[1].set_title("TEST")
```

- A maioria das features clínicas apresenta uma distribuição proporcionalmente semelhante entre treino e teste
- A exceção é a feature de "Fever"

```python
# 25 random Protein Features Distribution
feat_columns = [col for col in train.columns if "Freq" in col]
for i in np.random.randint(low=0, high=len(feat_columns), size=25):
    fig, ax = plt.subplots(ncols=2)
    _ = sns.boxplot(y=train[feat_columns[i]], ax=ax[0])
    _ = sns.boxplot(y=test[feat_columns[i]], ax=ax[1])
    max_value = max(train[feat_columns[i]].to_list() + test[feat_columns[i]].to_list())
    ax[0].set_ylim([0,max_value])
    ax[1].set_ylim([0,max_value])
    ax[0].set_title("TRAIN")
    ax[1].set_title("TEST")
```

- Checando as distribuições de algumas features selecionadas de forma aleatória para validar se não há casos de distribuições muito diferentes entre treino e teste (caso de "covariate shift")

- Porém os valores parecem estar em um range parecido e não foi detectada alguma variável com diferença significativa entre treino e teste

```python
# Count Cases 'Group' by 'Sex'
train.groupby(["Sex", "Group"])[["Age (Y)"]].count().rename(columns={"Age (Y)": "Count"})
```

```python
# Creating column for age group
train["age_group"] = ""
train.loc[train["Age (Y)"] <= 13, "age_group"] = "0 - 13"
train.loc[(train["Age (Y)"] > 13) & (train["Age (Y)"] <= 21), "age_group"] = "13 - 21"
train.loc[(train["Age (Y)"] > 21) & (train["Age (Y)"] <= 50), "age_group"] = "21 - 50"
train.loc[(train["Age (Y)"] > 50) & (train["Age (Y)"] <= 70), "age_group"] = "50 - 70"
train.loc[(train["Age (Y)"] > 70) & (train["Age (Y)"] <= 100), "age_group"] = "70 - 100"
```

```python
# Count Cases 'Group' by 'Sex'
train.groupby(["age_group", "Group"])[["Age (Y)"]].count().rename(columns={"Age (Y)": "Count"})
```

- Não existem casos severos para idades até 21 anos
- A maioria dos casos severos se encontra entre pacientes de 50 a 70 anos

```python
# Count Cases 'Group' by 'Sex'
train.groupby(["Sex", "age_group", "Group"])[["Age (Y)"]].count().rename(columns={"Age (Y)": "Count"})
```

- Combinando a informação de faixa etária e sexo, podemos ver que a maioria dos casos severos se encontra em pacientes do sexo masculino com idade entre 21 e 50 anos

- Para o sexo feminino a faixa etária com mais casos severos é a de 50 a 70 anos

```python
# Most common symptoms for severe cases
severe_symptoms = train[(train.Group=="SEVERE") & (~train.isna().any(1))][symptoms_cols].sum().to_frame()/np.sum(train.Group=="SEVERE")
severe_symptoms.rename(columns={0: "%_cases"}).sort_values("%_cases", ascending=False)
```

```python
# Most common symptoms for mild cases
mild_symptoms = train[(train.Group=="MILD") & (~train.isna().any(1))][symptoms_cols].sum().to_frame()/np.sum(train.Group=="MILD")
mild_symptoms.rename(columns={0: "%_cases"}).sort_values("%_cases", ascending=False)
```
